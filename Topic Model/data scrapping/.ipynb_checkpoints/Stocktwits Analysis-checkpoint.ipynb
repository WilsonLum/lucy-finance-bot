## Define Functions

```python
# Text preprocesing

import nltk
from nltk.corpus import stopwords
mystopwords=stopwords.words("English") + ['']
WNlemma = nltk.WordNetLemmatizer()

def pre_process(text):
    # Remove all the http links
    #processed_tweet = re.sub(r'/[\s-]+/', '', text)
    processed_tweet = re.sub(r'(https|http)?:\/\/(\w|\.|\/|\?|\=|\&|\%)*\b', '', text,flags=re.MULTILINE)
    # Remove all the special characters
    processed_tweet = re.sub(r'\W', ' ', processed_tweet)
    # Substituting multiple spaces with single space
    processed_tweet= re.sub(r'\s+', ' ', processed_tweet, flags=re.I)
    # Substituting rows with only numbers with null
    processed_tweet = re.sub(r'[0-9]+', '', text)

    tokens = nltk.word_tokenize(processed_tweet)
    #tokens=[ WNlemma.lemmatize(t.lower()) for t in tokens]
    tokens = [ t.lower() for t in tokens]
    tokens = [ t for t in tokens if t not in mystopwords]
    tokens = [ t for t in tokens if len(t) >= 3 ]
    text_after_process=" ".join(tokens)
    return(text_after_process)  
```

## Extract Sentiment & Subjectivity scores using textblob with Preprocessed Text

- The polarity score is a float within the range [-1.0, 1.0]. 
- The subjectivity is a float within the range [0.0, 1.0] where 0.0 is very objective and 1.0 is very subjective.

```python
stocks['text'] = stocks.body.apply(pre_process)
```

# Topic Model

```python
def topic_pre_process(text):
    tokens = nltk.word_tokenize(text)
    tokens=[ WNlemma.lemmatize(t.lower()) for t in tokens]
    tokens=[ t for t in tokens if t not in mystopwords]
    tokens = [ t for t in tokens if len(t) >= 3 ]
    return(tokens)
```

```python
toks = stocks['text'].apply(topic_pre_process)
```

```python
# Use dictionary (built from corpus) to prepare a DTM (using frequency)
import logging
import gensim 
from gensim import corpora

logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
```

```python
# Filter off any words with document frequency less than 3, or appearing in more than 80% documents
dictionary = corpora.Dictionary(toks)
print(dictionary)
dictionary.filter_extremes(no_below=3, no_above=0.7)
print(dictionary)

dtm = [dictionary.doc2bow(d) for d in toks]
```

```python
# Build an LDA model for 2 topics by default
lda_default = gensim.models.ldamodel.LdaModel(dtm, num_topics = 2,id2word = dictionary)
lda_default.show_topics(20)
```

```python
# Build an LDA model for 3 topics by default
lda_default = gensim.models.ldamodel.LdaModel(dtm, num_topics = 3,id2word = dictionary)
lda_default.show_topics(10)
```
