```python
import pandas as pd
```

## Reading in the twiter dataset

```python
f_bull = open("github data/bullish.txt","r")
f_bear = open("github data/bearish.txt","r")
```

```python
df_bull = pd.DataFrame(f_bull,columns = ['text'])
df_bear = pd.DataFrame(f_bear,columns = ['text'])
```

```python
df_bull
```

```python
df_bull['sentiment'] = 'Bullish'
df_bear['sentiment'] = 'Bearish'
```

```python
stocks = pd.concat([df_bull,df_bear])
```

```python
stocks
```

```python
len(stocks)
```

```python
stocks.sentiment.value_counts()
```

```python
#*******************************************************************************
#                         Save data
#*******************************************************************************
from datetime import date
# Month abbreviation, day and year
today = date.today()
d = today.strftime("%b-%d-%Y")

# Create a new excel workbook
writer = pd.ExcelWriter('data/twiters_label_data_' + d + '.xlsx', engine='xlsxwriter')

stocks.to_excel(writer, sheet_name= 'stocks_label', index=False)

# Close the Pandas Excel writer and output the Excel file.
writer.save()
```
